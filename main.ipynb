```python
import newspaper
import pandas as pd
```

```python
# Intailizing the givins
type_random = [86, 32, 73, 24, 76, 38, 45, 18, 44, 12]
demand_random = [4, 39, 66, 89, 97, 24, 9, 55, 15, 17]
order_quantity = 70
sell_price = 0.50
cost_price = 0.33
scrape = 0.05
news_types = ['good', 'fair', 'poor']
news_types_probs = [0.35, 0.45, 0.20]
demands_values = [40, 50, 60, 70, 80, 90, 100]
good_demand_probs = [0.03, 0.05, 0.15, 0.20, 0.35, 0.15, 0.07]
fair_demand_probs = [0.10, 0.18, 0.40, 0.20, 0.08, 0.04]
poor_demand_probs = [0.44, 0.22, 0.16, 0.12, 0.06]
```

```python
# Change the parameters for different givins for the simulation
data = newspaper.simulate(type_random, demand_random, order_quantity, sell_price,
                          cost_price, scrape, news_types, news_types_probs, 
                          demands_values, good_demand_probs, fair_demand_probs, 
                          poor_demand_probs, days=10)
```

```python
# calc total profit
print(f"{round(data['daily_profit'].sum(), 1)} " + "$")
```

```python
# write the output to csv file
data.to_csv("output.csv")
```
